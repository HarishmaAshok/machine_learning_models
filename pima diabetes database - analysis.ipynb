```python
import pandas as pd
```

```python
# importing required dataset - u.data
# cols = ['userId', 'movie id', 'rating', 'timestamp']
df = pd.read_csv(r'F:\Harshi College\MS DA\Machine Learning\lab\unsupervised learning\diabetes.csv',sep=',')
print(df.head)
print(df.columns.values)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['Outcome'], axis = 1)
y = df['Outcome']

training_data, testing_data, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)

print(f"No. of training examples: {training_data.shape[0]}")
print(f"No. of testing examples: {testing_data.shape[0]}")
```

```python
training_data
```

##### KNN classifier

```python
# KNN classifier

from datetime import datetime
now = datetime.now()
start_time = now.strftime("%H:%M:%S")
print("start time =", start_time)

accList = []
# Import necessary modules
from sklearn.neighbors import KNeighborsClassifier
for i in range(40,50):
    knn = KNeighborsClassifier(n_neighbors=i)

    knn.fit(training_data, y_train)

    # Predict on dataset which model has not seen before
    y_pred = knn.predict(testing_data)
#     print(y_pred)

    import numpy as np
    # converting series to ndarray
    y_test = np.asarray(y_test)

    # confusion matrix
    from sklearn.metrics import confusion_matrix
    tp_tn, fp_fn = confusion_matrix(y_test, y_pred)
#     print(confusion_matrix(y_test, y_pred))
    print(f"true positive : {list(tp_tn)[0]}\n true negative : {list(tp_tn)[1]}\n false positive : {list(fp_fn)[0]}\n false negative : {list(fp_fn)[1]}\n")

    # error evaluation methods
    # calculating accuracy

    from sklearn.metrics import classification_report
    from sklearn.metrics import accuracy_score

    #Check precision, recall, f1-score
    print(classification_report(y_test, y_pred) )

    print("accuracy_score",accuracy_score(y_test, y_pred))
    print("is with k = ",i)
    accList.append(accuracy_score(y_test, y_pred))
    print("=======================================================================================")

from statistics import mean
print("mean of the accuracy is")
mean(accList)


now = datetime.now()
end_time = now.strftime("%H:%M:%S")
print("end time =", end_time)

# computing time complexity
time_1 = datetime.strptime(start_time,"%H:%M:%S")
time_2 = datetime.strptime(end_time,"%H:%M:%S")

time_interval = time_2 - time_1
print(time_interval)
 
```

```python
# Decision tree

from datetime import datetime
now = datetime.now()
start_time = now.strftime("%H:%M:%S")
print("start time =", start_time)

from sklearn.tree import DecisionTreeClassifier
clf_model = DecisionTreeClassifier(criterion="gini", random_state=42, max_depth=3, min_samples_leaf=5)   
clf_model.fit(training_data,y_train)

y_pred = clf_model.predict(testing_data)
y_pred

# confusion matrix
from sklearn.metrics import confusion_matrix
tp_tn, fp_fn = confusion_matrix(y_test, y_pred)
#print(confusion_matrix(y_test, y_pred))
print(f"true positive : {list(tp_tn)[0]}\n true negative : {list(tp_tn)[1]}\n false positive : {list(fp_fn)[0]}\n false negative : {list(fp_fn)[1]}\n")

# error evaluation methods
# calculating accuracy

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

#Check precision, recall, f1-score
print(classification_report(y_test, y_pred) )

print("accuracy_score",accuracy_score(y_test, y_pred))


now = datetime.now()
end_time = now.strftime("%H:%M:%S")
print("end time =", end_time)

# computing time complexity
time_1 = datetime.strptime(start_time,"%H:%M:%S")
time_2 = datetime.strptime(end_time,"%H:%M:%S")

time_interval = time_2 - time_1
print(time_interval)
 
```
